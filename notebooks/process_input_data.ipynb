```python
import pandas as pd
```

```python
input_data = pd.read_csv("../data/features_50_filtered.csv")
input_data.head()
```

```python
pretrain_data = input_data.iloc[:, 2:].copy()
pretrain_data.columns = [f"X_{i}" for i in range(len(pretrain_data.columns))]
pretrain_data.to_csv("../data/pretrain.csv", index=False)
pretrain_data.head()
```

```python
train_data = input_data.sample(n=350).copy()
del train_data["month"]
train_data.columns = ["Y"] + [f"X_{i}" for i in range(len(train_data.columns) - 1)]
train_data.to_csv("../data/train.csv", index=False)
train_data.head()
```

```python
test_data = input_data.sample(n=150).copy()
del test_data["month"]
test_data.columns = ["Y"] + [f"X_{i}" for i in range(len(test_data.columns) - 1)]
test_data.to_csv("../data/test.csv", index=False)
test_data.head()
```
